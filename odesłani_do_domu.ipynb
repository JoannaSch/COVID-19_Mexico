```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_data_nonas.csv")
df
```

```python
#tabela osÃ³b odesÅ‚anych do domu
df_home = df[df['PATIENT_TYPE'] == 0]
df_home
```

```python
#tabela osÃ³b hospitalizowanych
df_hospital = df[df['PATIENT_TYPE'] == 1]
df_hospital
```

```python
pd.crosstab(df_home['died'], df_home['CLASIFFICATION_FINAL']) #podglÄ…d osÃ³b z tabeli hospitalizowanych w ujÄ™ciu zgonÃ³w i covida
```

```python
pd.crosstab(df_hospital['died'], df_hospital['CLASIFFICATION_FINAL']) #podglÄ…d osÃ³b z tabeli hospitalizowanych w ujÄ™ciu zgonÃ³w i covida
```

```python
#wyliczenie Å›miertelnoÅ›ci wÅ›rÃ³d osÃ³b wypuszczonych do domu
surv_home = len(df_home[df_home['died'] == 0])
print(surv_home)
died_home = len(df_home[df_home['died'] == 1])
print(died_home)
P_home = died_home / (died_home + surv_home)
P_home
```

```python
#wyliczenie Å›miertelnoÅ›ci wÅ›rÃ³d osÃ³b hospitalizowanych
surv_hospital = len(df_hospital[df_hospital['died'] == 0])
print(surv_hospital)
died_hospital = len(df_hospital[df_hospital['died'] == 1])
print(died_hospital)
P_hospital = died_hospital / (died_hospital + surv_hospital)
P_hospital
```

```python
#wyliczenie Å›miertelnoÅ›ci w caÅ‚ej prÃ³bie
surv_all = len(df[df['died'] == 0])
print(surv_all)
died_all = len(df[df['died'] == 1])
print(died_all)
P_died = died_all / (died_all + surv_all)
P_died
```

```python
#stworzenie sÅ‚ownika z wyliczonÄ… Å›miertelnoÅ›ciÄ…
cov_died_percentage = dict()
cov_died_percentage['home'] = P_home
cov_died_percentage['hospital'] = P_hospital
```

```python
died_names = list(cov_died_percentage.keys())
died_prob = list(cov_died_percentage.values())

#wykres Å›miertelnoÅ›ci wÅ›rÃ³d pacjentÃ³w odesÅ‚anych vs hospitalizowanych (nie tylko z potwierdzonym COVID-em) 
#+ ogÃ³lna Å›miertelnoÅ›Ä‡ w prÃ³bie jako linia referencyjna
plt.bar(range(len(died_names)), died_prob, tick_label=died_names)
plt.ylim(0, 0.5)
plt.xticks(rotation = 90)
plt.axhline(y=P_died, color= 'green', linewidth=2,)
plt.show()
```

```python
#tabela osÃ³b odesÅ‚anych do domu z covidem
df_home_covid = df[(df['PATIENT_TYPE'] == 0) & (df['CLASIFFICATION_FINAL'] == 1)]
df_home_covid
```

```python
#tabela osÃ³b hospitalizowanych z covidem
df_hospital_covid = df[(df['PATIENT_TYPE'] == 1) & (df['CLASIFFICATION_FINAL'] == 1)]
df_hospital
```

```python
#wyliczenie proporcji pacjentÃ³w z covidem wÅ›rÃ³d wypuszczonych
covid_home = len(df_home[df_home['CLASIFFICATION_FINAL'] == 1])
print(covid_home)
all_home = len(df_home)
print(all_home)
P_covid_home = covid_home / all_home
P_covid_home
```

```python
#wyliczenie proporcji pacjentÃ³w z covidem wÅ›rÃ³d hospitalizowanych
covid_hospital = len(df_hospital[df_hospital['CLASIFFICATION_FINAL'] == 1])
print(covid_hospital)
all_hospital = len(df_hospital)
print(all_hospital)
P_covid_hospital = covid_hospital / all_hospital
P_covid_hospital
```

```python
#wyliczenie zachorowalnoÅ›ci na covid w caÅ‚ej prÃ³bie
covid_all = len(df[df['CLASIFFICATION_FINAL'] == 1])
print(covid_all)
all_patients = len(df)
print(all_patients)
P_covid_all = covid_all / all_patients
P_covid_all
```

```python
#stworzenie sÅ‚ownika z zachorowalnoÅ›ciÄ…
cov_percentage = dict()
cov_percentage['home'] = P_covid_home
cov_percentage['hospital'] = P_covid_hospital
```

```python
covid_names = list(cov_percentage.keys())
covid_prob = list(cov_percentage.values())

#odestek chorych na COVID wÅ›rÃ³d pacjentÃ³w odesÅ‚anych vs hospitalizowanych + ogÃ³lny ich odsetek w prÃ³bie jako linia referencyjna
plt.bar(range(len(covid_names)), covid_prob, tick_label=covid_names)
plt.ylim(0, 0.7)
plt.xticks(rotation = 90)
plt.axhline(y=P_covid_all, color= 'green', linewidth=2,)
plt.show()
```

```python
#stworzenie sÅ‚ownika z procentem osÃ³b odesÅ‚anych do domu w kaÅ¼dej grupie wiekowej
home_age = dict()
home_age['0-20'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 1)]) /len(df['age_group'] == 1)
home_age['21-40'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 2)]) /len(df['age_group'] == 2)
home_age['41-60'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 3)]) /len(df['age_group'] == 3)
home_age['61-80'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 4)]) /len(df['age_group'] == 4)
home_age['81+'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 5)]) /len(df['age_group'] == 5)
```

```python
home_age_names = list(home_age.keys())
home_age_prob = list(home_age.values())

#proporcje osÃ³b odesÅ‚anych do domu w kaÅ¼dej grupie wiekowej
#(czyli dla kaÅ¼dej grupy wiekowej - ile procent pacjentÃ³w byÅ‚o odesÅ‚anych do domu - poszczegÃ³lnych sÅ‚upkÃ³w nie moÅ¼na sumowaÄ‡, bo to rÃ³Å¼ne grupy)
plt.bar(range(len(home_age_names)), home_age_prob, tick_label=home_age_names)
plt.ylim(0, 0.7)
plt.xticks(rotation = 90)
plt.show()
```

```python
#stworzenie sÅ‚ownika z procentem zmarÅ‚ych wÅ›rÃ³d osÃ³b odesÅ‚anych do domu w kaÅ¼dej grupie wiekowej
home_died_age = dict()
home_died_age['0-20'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 1) & (df['died'] == 1)]) /len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 1)])
home_died_age['21-40'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 2) & (df['died'] == 1)]) /len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 2)])
home_died_age['41-60'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 3) & (df['died'] == 1)]) /len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 3)])
home_died_age['61-80'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 4) & (df['died'] == 1)]) /len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 4)])
home_died_age['81+'] = len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 5) & (df['died'] == 1)]) /len(df[(df['PATIENT_TYPE'] == 0) & (df['age_group'] == 5)])
```

```python
home_died_age_names = list(home_died_age.keys())
home_died_age_prob = list(home_died_age.values())

#proporcje osÃ³b zmarÅ‚ych wÅ›rÃ³d odesÅ‚anych do domu w kaÅ¼dej grupie wiekowej
plt.bar(range(len(home_died_age_names)), home_died_age_prob, tick_label=home_died_age_names)
plt.ylim(0, 0.1)
plt.xticks(rotation = 90)
plt.axhline(y=P_home, color= 'green', linewidth=2,)
plt.show()
```
