```python
import pandas as pd
```

```python
df = pd.read_csv("Covid Data.csv")
df
```

```python
df['died'] = df['DATE_DIED']
df
```

```python
df.loc[df['died'] == '9999-99-99', 'died'] = 0
df
```

```python
df.loc[df['died'] != 0, 'died'] = 1
df
```

```python
df['died'] = df['died'].astype('int64')
```

```python
df.dtypes
```

```python
df.loc[df['CLASIFFICATION_FINAL'] >= 4, 'CLASIFFICATION_FINAL'] = 0
df
```

```python
df.loc[df['CLASIFFICATION_FINAL'] > 0, 'CLASIFFICATION_FINAL'] = 1
df
```

```python
df.loc[df['PATIENT_TYPE'] == 1, 'PATIENT_TYPE'] = 0
df.loc[df['PATIENT_TYPE'] == 2, 'PATIENT_TYPE'] = 1
df
```
